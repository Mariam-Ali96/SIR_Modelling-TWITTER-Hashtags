```python
file_name1= '/home/mariam/Downloads/DATA_for_visualization/data_v1.dat'
file_name2= '/home/mariam/Downloads/DATA_for_visualization/data_v2.dat'
file_name3= '/home/mariam/Downloads/DATA_for_visualization/data_v3.dat'
file_name4= '/home/mariam/Downloads/DATA_for_visualization/data_v4.dat'
file_name5= '/home/mariam/Downloads/DATA_for_visualization/data_v5.dat'


color_1= '/home/mariam/Downloads/DATA_for_visualization/color_v1.dat'
color_2= '/home/mariam/Downloads/DATA_for_visualization/color_v2.dat'
color_3= '/home/mariam/Downloads/DATA_for_visualization/color_v3.dat'
color_4= '/home/mariam/Downloads/DATA_for_visualization/color_v4.dat'
color_5= '/home/mariam/Downloads/DATA_for_visualization/color_v5.dat'





M = pd.read_csv(file_name1, sep="\t", header=None)
N = pd.read_csv(file_name2, sep="\t", header=None)
P = pd.read_csv(file_name3, sep="\t", header=None)
O = pd.read_csv(file_name4, sep="\t", header=None)
Q= pd.read_csv(file_name5, sep="\t", header=None)

y_1 = pd.read_csv(color_1, sep="\t", header=None)
y_2 = pd.read_csv(color_1, sep="\t", header=None)
y_3 = pd.read_csv(color_3, sep="\t", header=None)
y_4 = pd.read_csv(color_4, sep="\t", header=None)
y_5 = pd.read_csv(color_5, sep="\t", header=None)


y_1 = np.array(y_1).reshape(1000,)
y_2 = np.array(y_2).reshape(1000,)
y_3 = np.array(y_3).reshape(1000,)
y_4 = np.array(y_4).reshape(1000,)

y_5 = np.array(y_5).reshape(1000,)


```

```python
#threshold and splic colors
y_10 = np.where(y_1 > 0.5, 1, 0)
y_20 = np.where(y_2 > 0.5, 1, 0)
y_30 = np.where(y_3 > 0.5, 1, 0)
y_40 = np.where(y_4 > 0.5, 1, 0)
y_50 = np.where(y_5 > 0.5, 1, 0)
```

```python
#colors definitions
red_1 = y_10 == 0
green_1 = y_10 == 1




blue = y_20 == 0
cyan = y_20 == 1



mag = y_30 == 0
yellow = y_30 == 1


black = y_40 == 0
white = y_40 == 1

red_5 = y_50 == 0
green_5 = y_50 == 1
```

# Plotting the first dataset M according to coloring y_1

```python
#Normaliz M dataset


from sklearn import preprocessing

x = M.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
M_new = pd.DataFrame(x_scaled)
x_scaled = x_scaled.reshape(2500,2)
```

```python
n_samples = 1000
n_components = 3
(fig, subplots) = plt.subplots(3, 5, figsize=(15, 8))
perplexities = [5, 30, 50, 100]


ax = subplots[0][0]
ax.scatter(M.iloc[red_1, 0], M.iloc[red_1, 1], c="r")
ax.scatter(M.iloc[green_1, 0], M.iloc[green_1, 1], c="g")
ax.xaxis.set_major_formatter(NullFormatter())
ax.yaxis.set_major_formatter(NullFormatter())
plt.axis('tight')

for i, perplexity in enumerate(perplexities):
    ax = subplots[0][i + 1]

    t0 = time()
    tsne = manifold.TSNE(n_components=2, init='random',
                         random_state=0, perplexity=perplexity)
    Mx1 = tsne.fit_transform(M)
    t1 = time()
    print("circles, perplexity=%d in %.2g sec" % (perplexity, t1 - t0))
    ax.set_title("Perplexity=%d" % perplexity)
    ax.scatter(Mx1[red_1, 0], Mx1[red_1, 1], c="r")
    ax.scatter(Mx1[green_1, 0], Mx1[green_1, 1], c="g")
    ax.xaxis.set_major_formatter(NullFormatter())
    ax.yaxis.set_major_formatter(NullFormatter())
    ax.axis('tight')

# Another example using s-curve
#X, color = datasets.samples_generator.make_s_curve(n_samples, random_state=0)

ax = subplots[1][0]
ax.scatter(M.iloc[:, 0], M.iloc[:, 2], c=np.array(y_1).reshape(1000,))
ax.xaxis.set_major_formatter(NullFormatter())
ax.yaxis.set_major_formatter(NullFormatter())

for i, perplexity in enumerate(perplexities):
    ax = subplots[1][i + 1]

    t0 = time()
    tsne = manifold.TSNE(n_components=3, init='random',
                         random_state=0, perplexity=perplexity)
    My1 = tsne.fit_transform(M)
    t1 = time()
    print("S-curve, perplexity=%d in %.2g sec" % (perplexity, t1 - t0))

    ax.set_title("Perplexity=%d" % perplexity)
    ax.scatter(My1[:, 0], My1[:, 1], c=np.array(y_1).reshape(1000,))
    ax.xaxis.set_major_formatter(NullFormatter())
    ax.yaxis.set_major_formatter(NullFormatter())
    ax.axis('tight')
plt.show()
```
